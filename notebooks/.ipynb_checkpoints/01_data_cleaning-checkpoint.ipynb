```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

print("All libraries imported successfully!")
```

```python
df = pd.read_csv('../data/SampleSuperstore.csv', encoding='latin1')
print("Shape:", df.shape)
print("\nFirst 5 rows:")
df.head()
```

```python
print("Column Names:")
print(df.columns.tolist())

print("\nData Types:")
print(df.dtypes)

print("\nBasic Statistics:")
df.describe()
```

```python
print("Missing Values in each column:")
print(df.isnull().sum())

print("\nTotal Duplicate Rows:", df.duplicated().sum())

print("\nTotal Rows:", len(df))
```

```python
# Remove duplicate rows
df = df.drop_duplicates()

print("Rows after removing duplicates:", len(df))
print("Duplicates removed:", 9994 - len(df))
```

```python
# Profit Margin per order
df['Profit_Margin'] = round((df['Profit'] / df['Sales']) * 100, 2)

# Discount Bands (same as our SQL query)
df['Discount_Band'] = pd.cut(
    df['Discount'],
    bins=[-1, 0, 0.20, 0.40, 1.0],
    labels=['No Discount', 'Low (1-20%)', 'Medium (21-40%)', 'High (41%+)']
)

print("New columns added successfully!")
print(df[['Sales', 'Profit', 'Profit_Margin', 'Discount', 'Discount_Band']].head())
```

```python
df.to_csv('../data/superstore_clean.csv', index=False)
print("Cleaned file saved successfully!")
```
